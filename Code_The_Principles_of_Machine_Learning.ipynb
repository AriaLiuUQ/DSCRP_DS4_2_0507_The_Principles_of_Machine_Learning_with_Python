Missing Data Excercise

```python
!pip install missingno
import pandas as pd

# Load dataset directly from URL
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00492/Metro_Interstate_Traffic_Volume.csv.gz"
df_1 = pd.read_csv(url)

# Print data
print(df_1.head())
print(df_1.info())
print(df_1.isnull().sum())

```

```python
import missingno as msno
import matplotlib.pyplot as plt

# Visualize missing values BEFORE cleaning
print("Missing values BEFORE cleaning:")
print(df_1.isnull().sum())

msno.matrix(df_1)
plt.title("Missing Values (Before Cleaning)")
plt.show()

msno.bar(df_1)
plt.title("Missing Value Count (Before Cleaning)")
plt.show()
```

```python
# Numeric columns
for col in df_1.select_dtypes(include=['float64', 'int64']):
    df_1[col] = df_1[col].fillna(df_1[col].mean())

# Categorical columns
for col in df_1.select_dtypes(include=['object']):
    df_1[col] = df_1[col].fillna(df_1[col].mode()[0])

# Convert date column​
df_1['date_time'] = pd.to_datetime(df_1['date_time'])

# Remove duplicates​
df_1 = df_1.drop_duplicates()
```

```python
# Visualize missing values AFTER cleaning
print("Missing values AFTER cleaning:")
print(df_1.isnull().sum())

msno.matrix(df_1)
plt.title("Missing Values (After Cleaning)")
plt.show()

msno.bar(df_1)
plt.title("Missing Value Count (After Cleaning)")
plt.show()
```

Data Transformation


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import seaborn as sns

plt.figure(figsize=(14,6))
sns.boxplot(data=X_scaled)
plt.xticks(rotation=90)
plt.title("Feature Distribution (Standardized) - Breast Cancer Dataset")
plt.show()
```

Train and test set split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
y_train = y_train.values.ravel()  # reshape y_train to 1D array
y_test = y_test.values.ravel()   # reshape y_test to 1D array
```

KNN model

```python
#K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
# Predict on the test set
y_pred_knn = knn.predict(X_test)
```

Print KNN model evaluation

```python
# Evaluate the model - accuracy
from sklearn.metrics import accuracy_score
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"KNN Accuracy: {accuracy_knn:.2f}")
```

Confusion Matrix

```python

from sklearn.metrics import classification_report, confusion_matrix
print("KNN Classification Report:")
print(classification_report(y_test, y_pred_knn))
print("KNN Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))
```

K-fold Cross validation

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Define KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Regular K-Fold CV
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Cross-validation scores
kf_scores = cross_val_score(knn, X_scaled, y_train, cv=kf, scoring='accuracy')

print("K-Fold CV scores:", kf_scores)
print("Mean accuracy:", np.mean(kf_scores))
```

Stratified K-Fold CV

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Define KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Stratified 5-fold CV
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(knn, X_scaled, y_train, cv=skf, scoring='accuracy')

print("Cross-validation scores:", scores)
print("Mean CV accuracy:", np.mean(scores))
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Predict
y_pred = log_reg.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

rf_model = RandomForestClassifier()

# Train
rf_model.fit(X_train, y_train)

# Predict
y_pred = rf_model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
```
