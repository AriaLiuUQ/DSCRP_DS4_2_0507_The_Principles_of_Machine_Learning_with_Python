Data Encoding

```python
import sklearn 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Diagnosis'] = le.fit_transform(df['Diagnosis'])
print(df['Diagnosis'].value_counts())
```

Data Transformation


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Train and test set split 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
y_train = y_train.values.ravel()  # reshape y_train to 1D array
y_test = y_test.values.ravel()   # reshape y_test to 1D array
```

KNN model

```python
#K-Nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
# Predict on the test set
y_pred_knn = knn.predict(X_test)
```

Print KNN model evaluation 

```python
# Evaluate the model - accuracy
from sklearn.metrics import accuracy_score
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"KNN Accuracy: {accuracy_knn:.2f}")
```

Confusion Matrix 

```python

from sklearn.metrics import classification_report, confusion_matrix
print("KNN Classification Report:")
print(classification_report(y_test, y_pred_knn))
print("KNN Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_knn))
```

K-fold Cross validation

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Define KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Regular K-Fold CV
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Cross-validation scores
kf_scores = cross_val_score(knn, X_scaled, y_train, cv=kf, scoring='accuracy')

print("K-Fold CV scores:", kf_scores)
print("Mean accuracy:", np.mean(kf_scores))
```

Stratified K-Fold CV

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)

# Define KNN
knn = KNeighborsClassifier(n_neighbors=5)

# Stratified 5-fold CV
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(knn, X_scaled, y_train, cv=skf, scoring='accuracy')

print("Cross-validation scores:", scores)
print("Mean CV accuracy:", np.mean(scores))
```

Grid Search for KNN

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

# --- Step 1: Encode y ---
le = LabelEncoder()
y_encoded = le.fit_transform(y)  # e.g., 'B' -> 0, 'M' -> 1, etc.

# --- Step 2: Scale X ---
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)  # X and y must match in rows

# --- Step 3: Define KNN and param grid ---
knn = KNeighborsClassifier()
param_grid = {
    "n_neighbors": [3, 5, 7, 9, 11],
    "metric": ["euclidean", "manhattan"]
}

# --- Step 4: GridSearchCV ---
grid_search = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy"
)

# --- Step 5: Fit ---
grid_search.fit(X_scaled, y_encoded)

# --- Step 6: Results ---
print("Best Parameters:", grid_search.best_params_)
print("Best CV Score:", grid_search.best_score_)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Predict
y_pred = log_reg.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
